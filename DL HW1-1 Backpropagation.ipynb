## Q1 Q2
- backpropagation實作
- 比較zero and random initializations

```python
def init_params(num_x, num_hl, num_h, num_y):
    """
    num_x : input的維度
    num_hl : hidden layer的數量
    num_h : hidden layer的維度(list)
    num_y : 輸出的維度
    W1 : input -> hidden1
    WO : hidden last -> output
    """
    params={}
    np.random.seed(42)
    
    for l in range(2,num_hl+1):
        params['W%s'%l]=np.random.normal(0,1,(num_h[l-1], num_h[l-2]))*0.01 #params['Wi']=[(連結第一個hidden neuron的權重們(w11,w21,...)),(連結第二個hidden neuron的權重們(w12,w22,...))...]
#         params['W%s'%l]=np.zeros((num_h[l-1], num_h[l-2])) 
        params['b%s'%l]=np.zeros((num_h[l-1],1))
    
    
    params['W1']=np.random.normal(0,1,(num_h[0], num_x))*0.01
#     params['W1']=np.zeros((num_h[0], num_x))
    params['b1']=np.zeros((num_h[0], 1))
    params['WO']=np.random.normal(0,1,(num_y, num_h[-1]))*0.01
#     params['WO']=np.zeros((num_y, num_h[-1]))
    params['bO']=np.zeros((num_y,1))
    
    # remark z1 = (params['W1'].dot(input)+params['b1'])
    return params
```

```python
def FP(inputs, params):
    value=[]
    temp={}
    A=inputs
    
    def Relu(value):
        return np.maximum(0,value)
    
    def Softmax(value):    
        value=value-value.max(axis=0)
        exp=np.exp(value)     
        return exp/np.sum(exp,axis=0)
    
    for i in range(1,int(len(params)/2)):
        temp['W']=params['W%s'%i]
        temp['b']=params['b%s'%i]
        temp['A']=A #inputs

        temp['Z']=temp['W'].dot(temp['A'])+temp['b'] #ouputs befor activation func.
        
        A=Relu(temp['Z']) #這次output=下次input
        value.append(temp)
        temp={}
        
    temp['W']=params['WO']
    temp['b']=params['bO']
    temp['A']=A 
    temp['Z']=params['WO'].dot(A)+params['bO']
    
    value.append(temp)
    pred=Softmax(temp['Z'])
    
    return pred,value
```

```python
def BP(value, pred, true_Y, lr):
    grad=[]
    temp={}
    
    temp['dZ']=pred-true_Y
    temp['dW']=lr*np.dot(temp['dZ'],value[-1]['A'].T)  #dot時，就把不同sample的gradient加起來了
    temp['db']=np.sum(lr*temp['dZ'], axis=1, keepdims=True)/true_Y.shape[1]
    
    grad.append(temp)
    for i in range(2,len(value)+1):
        temp={}
        dA=np.dot(value[-(i-1)]['W'].T,grad[-(i-1)]['dZ']) #這層的dA=下面一層的dZ*下面一層的W
        dA[value[-i]['Z']<=0]=0 #relu微分，relu的input>0=1 else 0
        dZ=dA
        temp['dZ']=dZ
        temp['dW']=lr*np.dot(dZ,value[-i]['A'].T)
        temp['db']=np.sum(lr*dZ, axis=1, keepdims=True)/true_Y.shape[1]
        
        grad.insert(0,temp)
    
    return grad
```

```python
train=np.load(r'C:\Users\danie\OneDrive\桌面\DL HW1\Q1\train.npz')
train_y=np.asarray(pd.get_dummies(train['label'])).T
train_x=train['image'].reshape(12000,-1).T
train_x/=255
```

```python
test=np.load(r'C:\Users\danie\OneDrive\桌面\DL HW1\Q1\test.npz')
test_y=np.asarray(pd.get_dummies(test['label'])).T
test_x=test['image'].reshape(5768,-1).T
test_x/=255
```

```python
train_loss=[]
train_error=[]
test_error=[]

params=init_params(784,2,[256,128],10)
epoch=range(100)
batch=100

# train with batch size 100
for j in epoch:
    for i in range(0,12000,batch):
        pred, values=FP(train_x[:,i:i+batch], params)
        grad=BP(values, pred, train_y[:,i:i+batch],0.001)
    
        params['W1']-=grad[0]['dW']
        params['b1']-=grad[0]['db']
        params['W2']-=grad[1]['dW']
        params['b2']-=grad[1]['db']
        params['WO']-=grad[2]['dW']
        params['bO']-=grad[2]['db']
    
    pred_train, values=FP(train_x, params)
    train_loss.append(-(np.log(pred_train)*train_y).sum()/12000)
    train_error.append(1-sum(np.argmax(pred_train,axis=0)==train['label'])/12000)
    pred_test, values=FP(test_x, params)
    test_error.append(1-sum(np.argmax(pred_test,axis=0)==test['label'])/5768)
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1, 3, 1)
plt.xlabel('epochs'); plt.ylabel('error_rate')
plt.title('test_error_rate');
plt.plot(epoch,test_error)

plt.subplot(1, 3, 2)
plt.xlabel('epochs'); plt.ylabel('error_rate')
plt.title('train_error_rate');
plt.plot(epoch,train_error)

plt.subplot(1, 3, 3)
plt.xlabel('epochs'); plt.ylabel('cross_entropy')
plt.title('test_loss');
plt.plot(epoch,train_loss)
```

## Q4
- 分類的成效(Confusion Matrix)

```python
pred_test, values=FP(test_x, params)
pred_test=np.argmax(pred_test, axis=0)
test_y=np.int_(test['label'])
print('accuracy:',sum(test_y==pred_test)/len(test_y))
pd.crosstab(test_y, pred_test, rownames=['true'], colnames=['pred'])
```
