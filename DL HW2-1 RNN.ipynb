## Recurrent Neural Network for Classication
- 建立simple RNN及LSTM模型
- 用過去7天確診人數預測明天確診人數是否上升

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
from torch import nn
import pygal
import pygal_maps_world
from sklearn.utils import shuffle
```

```python
covid=pd.read_csv(r'C:\Users\danie\OneDrive\桌面\DL HW2\covid_19.csv',index_col=0)
```

```python
covid=covid-covid.shift(1,axis=1)
covid=covid.dropna(axis=1)
```

```python
corr=covid.transpose().corr()
```

```python
plt.figure(figsize = (15,15))
sns.heatmap(corr,mask=np.triu(corr),cbar_kws={"shrink": .5})  #np.triu(corr) give the upper triangle of the matrix
```

```python
#挑選相關係數大的國家來做 

country=[]
for i in range(185):
    for j in range(185): 
        if i!=j:
            if corr.iloc[i,j]>0.8:
                country.append(corr.index[i])
                country.append(corr.columns[j])
covid_train=covid.loc[set(country),]
len(covid_train)
```

```python
# 選3月後的data
covid_train=covid_train.iloc[:,38:]
covid_train=covid_train.T
label=(covid_train>0).astype(int)
```

```python
time_step=7
data=[]
for country in range(covid_train.shape[1]):
    for i in range(covid_train.shape[0]-time_step): 
        temp_x=torch.tensor(covid_train.iloc[0+i:time_step+i, country],dtype=torch.float32).reshape(-1,1)
        temp_y=torch.tensor(label.iloc[time_step+i, country],dtype=torch.float32).reshape(-1,1)
        data.append([temp_x, temp_y])

data=shuffle(data, random_state=42)

trainloader = torch.utils.data.DataLoader(data[:int(len(data)*0.8)], batch_size=10, num_workers=1, shuffle=True, drop_last=True)
testloader = torch.utils.data.DataLoader(data[int(len(data)*0.8):], batch_size=10, num_workers=1, shuffle=True, drop_last=True)
```

```python
#建立RNN

class RNN(nn.Module):
    def __init__(self):
        super(RNN, self).__init__()

        self.rnn = nn.RNN(  
            input_size=1,      #input的維度, 此例每個hidden input就一個數字
            hidden_size=32,    #幾個hidden unit
            num_layers=1,      #hidden 有幾層
            batch_first=True,  #batch會放在第一個維度 -> (batch, time_step, input_size)
            )
        
        self.out = nn.Linear(32, 1) #每個hidden32維,透過一個Linear後變成1維(因為label也是1個數字，這邊也可以設並變成其他維度)

    def forward(self, x, h_state): 
        # x (batch, time_step, input_size) 10*7*1
        # h_state (num_layers, batch, hidden_size) 1*7*32
        # r_out (batch, time_step, output_size) 10*7*32
        
        r_out, h_state = self.rnn(x, h_state) 
        outs = self.out(r_out[:,-1,:]) #拿output最後一個,因為我們只要一個output  10*1*32->10*1
        outs = outs.view(-1,1,1)  #reshape成跟labels一樣dimension
        outs = torch.sigmoid(outs) #轉到0~1之間才能算cross_entropy
        return outs, h_state


rnn = RNN()
print(rnn)
```

```python
#訓練模型

optimizer = torch.optim.Adam(rnn.parameters())   # optimize all rnn parameters
loss_func = nn.BCELoss()
h_state = None

train_acc=[]
test_acc=[]
for epoch in range(10):
    training_loss=0
    testing_loss=0
    train_pred=[]
    test_pred=[]

    for i,data in enumerate(trainloader, 0):
        inputs, labels = data
        prediction, h_state = rnn(inputs, h_state)   # rnn的inputs(x, h(t-1))
        h_state = h_state.data 
        
        loss = loss_func(prediction, labels)     # cross entropy loss
        training_loss+=loss
        optimizer.zero_grad()               # clear gradients for this training step
        loss.backward()                     # backpropagation, compute gradients
        optimizer.step()                    # update gradient
        train_pred.append(sum(torch.round(prediction)==labels)[0][0].item()/10)
    
    with torch.no_grad():
        for i,data in enumerate(testloader, 0):
            inputs, labels = data
            prediction, h_state = rnn(inputs,h_state)  
            loss = loss_func(prediction, labels)     
            testing_loss+=loss
            test_pred.append(sum(torch.round(prediction)==labels)[0][0].item()/10)
            
    train_acc.append(sum(train_pred)/len(train_pred))       
    test_acc.append(sum(test_pred)/len(test_pred))
    
    print('training_loss : %s'%(training_loss.item()/(i+1)), '  testing_loss: %s'%(testing_loss.item()/(i+1)))
    print('train_acc : %s'%(sum(train_pred)/len(train_pred)), '  test_acc : %s'%(sum(test_pred)/len(test_pred)), '\n')
```

```python
plt.figure(figsize = (12,3))
plt.subplot(1,2,1)
plt.xlabel('epochs'); plt.ylabel('accuracy');
plt.title('train');
plt.plot(range(1,11),train_acc,label='train')
plt.ylim(0.88,0.95)
plt.xticks(np.arange(1,11,1))
plt.legend()

plt.subplot(1,2,2)
plt.xlabel('epochs'); plt.ylabel('accuracy');
plt.title('test');
plt.plot(range(1,11),test_acc,label='test')
plt.ylim(0.88,0.95)
plt.xticks(np.arange(1,11,1))
plt.legend()
```

```python
#建立模型

class LSTM(nn.Module):
    def __init__(self):
        super(LSTM, self).__init__()

        self.lstm = nn.LSTM(  
            input_size=1,      #input的維度, 此例每個hidden input就一個數字
            hidden_size=32,    #幾個hidden unit
            num_layers=1,      #hidden 有幾層
            batch_first=True,  #batch會放在第一個維度 -> (batch, time_step, input_size)
            )
        
        self.out = nn.Linear(32, 1) #每個hidden32維,透過一個Linear後變成1維(因為label也是1個數字，這邊也可以設並變成其他維度)
    
    def forward(self, x): 
        # x (batch, time_step, input_size)
        # h_state (num_layers, batch, hidden_size)
        # r_out (batch, time_step, output_size)
        #torch.randn(1,10,32),torch.randn(1,10,32)
        r_out, _ = self.lstm(x,None) 
        outs = self.out(r_out[:,-1,:]) #拿output最後一個,因為我們只要一個output
        outs = outs.view(-1,1,1)  #reshape成跟labels一樣dimension
        outs = torch.sigmoid(outs) #轉道0~1之間才能算cross_entropy
        return outs


lstm = LSTM()
print(lstm)
```

```python
#訓練模型 

optimizer = torch.optim.Adam(lstm.parameters())   # optimize all rnn parameters
loss_func = nn.BCELoss()

train_acc=[]
test_acc=[]
for epoch in range(10):
    training_loss=0
    testing_loss=0
    train_pred=[]
    test_pred=[]
    for i,data in enumerate(trainloader, 0):
        inputs, labels = data
        prediction = lstm(inputs)   
        
        loss = loss_func(prediction, labels)     # cross entropy loss
        training_loss+=loss
        optimizer.zero_grad()               # clear gradients for this training step
        loss.backward()                     # backpropagation, compute gradients
        optimizer.step()                    # update gradient
        train_pred.append(sum(torch.round(prediction)==labels)[0][0].item()/len(prediction))
    
    with torch.no_grad():
        for i,data in enumerate(testloader, 0):
            inputs, labels = data
            prediction = lstm(inputs)   
            loss = loss_func(prediction, labels)    
            testing_loss+=loss
            test_pred.append(sum(torch.round(prediction)==labels)[0][0].item()/10)
     
    train_acc.append(sum(train_pred)/len(train_pred))       
    test_acc.append(sum(test_pred)/len(test_pred))
    print('training_loss : %s'%(training_loss.item()/(i+1)), '  testing_loss: %s'%(testing_loss.item()/(i+1)))
    print('train_acc : %s'%(sum(train_pred)/len(train_pred)), '  test_acc : %s'%(sum(test_pred)/len(test_pred)), '\n')
```

```python
plt.figure(figsize = (12,3))
plt.subplot(1,2,1)
plt.xlabel('epochs'); plt.ylabel('accuracy');
plt.title('train');
plt.plot(range(1,11),train_acc,label='train')
plt.ylim(0.88,0.95)
plt.xticks(np.arange(1,11,1))
plt.legend()

plt.subplot(1,2,2)
plt.xlabel('epochs'); plt.ylabel('accuracy');
plt.title('test');
plt.plot(range(1,11),test_acc,label='test')
plt.ylim(0.88,0.95)
plt.xticks(np.arange(1,11,1))
plt.legend()
```
